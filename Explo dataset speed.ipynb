```python
# Function to calculte distance
def calculate_dist (lat_1,lat_2,lon_1,lon_2):
    R_aver = 6374;
    lat_1 = math.radians(lat_1)
    lon_1 = math.radians(lon_1)
    lat_2 = math.radians(lat_2)
    lon_2 = math.radians(lon_2)
    if (math.isnan(lat_1)|math.isnan(lat_2)|math.isnan(lon_1)|math.isnan(lon_2)):
         return float('nan')         
    elif ((lat_1==lat_2)&(lon_1==lon_2)):
        dist=0.0
    elif (math.cos(lat_1)*math.cos(lat_2)*math.cos(lon_1-lon_2) + math.sin(lat_1)*math.sin(lat_2)) > 1:
        dist=0.0
    else:
        dist = R_aver * math.acos(math.cos(lat_1)*math.cos(lat_2)*math.cos(lon_1-lon_2) + math.sin(lat_1)*math.sin(lat_2));
    return dist*1000
```

```python
def calc_speed(time_start,time_end,distance):
    delta = parse(time_end) - parse(time_start)
    delta = delta.total_seconds()
    if delta == 0.0:
        a = 0
    else:
        a = (distance / delta) * 3.6
    return a
```

```python
#db_group = db_test.groupby(['PROBE_ID'])
def distance_speed(db_global):
    db_global = db_global.dropna()
    db_global = db_global.set_index(['PROBE_ID','SAMPLE_DATE']).sort_index()
    db_global.reset_index(inplace = True)

    lon_list = db_global['LON'].tolist()
    lon_list.insert(0, 0)

    lat_list = db_global['LAT'].tolist()
    lat_list.insert(0, 0)

    db_global.loc[:,'lon_list'] = lon_list[0:-1]
    db_global.loc[:,'lat_list'] = lat_list[0:-1]

    db_global.loc[1:,'distance'] = db_global.apply(lambda row: calculate_dist(row['lat_list'],row['LAT'],row['lon_list'],row['LON']),axis=1)

    SAMPLE_DATE_list = db_global['SAMPLE_DATE'].tolist()
    SAMPLE_DATE_list.insert(0, 0)

    db_global.loc[:,'SAMPLE_DATE_list'] = SAMPLE_DATE_list[0:-1]
    SAMPLE_DATE_list[0] = db_global['SAMPLE_DATE'][0]

    db_global['SAMPLE_DATE_list'][0] = db_global['SAMPLE_DATE'][0]

    db_global.loc[1:,'speed'] = db_global.apply(lambda row: calc_speed(row['SAMPLE_DATE_list'],row['SAMPLE_DATE'],row['distance']),axis=1)
    db_global = db_global.drop(columns=['SYSTEM_DATE', 'lon_list', 'lat_list', 'distance'])
    return db_global
```
