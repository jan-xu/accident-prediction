### Plot data points

```python
def plot_gps(df):
    plt.figure(figsize = (20,8))
    plt.scatter(df.LON, df.LAT, s = 0.5, c = df.SPEED)
    plt.xlim((-118.28, -118.2)) # to plot Downtown LA
    plt.ylim((34.03, 34.11))    # to plot Downtown LA
    #plt.axis('equal')
    plt.colorbar()
    plt.show()
    return
```

```python
plot_gps(df)
```
